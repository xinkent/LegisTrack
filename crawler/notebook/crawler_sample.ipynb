```python
import requests
from bs4 import BeautifulSoup
```

## 議案一覧の取得

```python
url = 'https://www.shugiin.go.jp/internet/itdb_gian.nsf/html/gian/kaiji213.htm'
```

```python
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
soup
```

```python
low_tables = soup.find_all("table")
```

```python
type(table)
```

```python
for table in low_tables:
    caption = table.find("caption").text
    if (caption == "衆法の一覧"):
        representative_low_table = table
```

```python
trs = representative_low_table.find_all("tr")

# 1行目はヘッダー。期待通りの列順でなければ例外。
header_tr = trs[0]

header = [td.text for td in header_tr.find_all("th")]
assert(header == ['提出回次', '番号', '議案件名', '審議状況', '経過情報', '本文情報'])
```

```python
for tr in trs[1:]:
    for i, td in enumerate(tr.find_all("td")):
        print("{}:{}".format(header[i], td.text))
        if td.text=="経過":
            print("link:{}".format(td.find("a").get("href")))
```

## 経過情報の取得

```python
url = 'https://www.shugiin.go.jp/internet/itdb_gian.nsf/html/gian/keika/1DDAD6A.htm'
```

```python
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
tables = soup.find_all("table")
keika_table = tables[0]
```

```python
for tr in keika_table.find_all("tr")[1:]:
    tds = tr.find_all("td")
    print("項目:{}".format(tds[0].text))
    print("内容:{}".format(tds[1].text.strip()))
```
